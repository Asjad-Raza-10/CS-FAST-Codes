### Question 1

```python
game_tree = [[[3, 5], [6, 9]], [[0, 1], [2, -1]]]
print(game_tree)
```

```python
import math

def alphabeta(node, depth, alpha, beta, is_maximizing_player):
    print(f"Depth {depth}: Current alpha={alpha}, beta={beta}, Player is {'MAX' if is_maximizing_player else 'MIN'}")

    # Base case: If node is a terminal node, return its value
    if not isinstance(node, list):
        print(f"Depth {depth}: Terminal node value = {node}")
        return node

    # Max node
    if is_maximizing_player:
        max_eval = -math.inf
        for i, child in enumerate(node):
            print(f"Depth {depth}: MAX player considering child {i+1} of {len(node)}")
            eval = alphabeta(child, depth + 1, alpha, beta, False)
            print(f"Depth {depth}: MAX player - Child {i+1} evaluated to {eval}. Current alpha={alpha}, beta={beta}")
            if eval > max_eval:
                max_eval = eval
            if max_eval > alpha:
                alpha = max_eval
                print(f"Depth {depth}: MAX player - Alpha updated to {alpha} (beta={beta})")

            if beta <= alpha:
                print(f"Depth {depth}: PRUNING - beta ({beta}) <= alpha ({alpha}). Pruning remaining branches.")
                break # Alpha-Beta Pruning
        return max_eval
    # Min node
    else:
        min_eval = math.inf
        for i, child in enumerate(node):
            print(f"Depth {depth}: MIN player considering child {i+1} of {len(node)}")
            eval = alphabeta(child, depth + 1, alpha, beta, True)
            print(f"Depth {depth}: MIN player - Child {i+1} evaluated to {eval}. Current alpha={alpha}, beta={beta}")
            if eval < min_eval:
                min_eval = eval
            if min_eval < beta:
                beta = min_eval
                print(f"Depth {depth}: MIN player - Beta updated to {beta} (alpha={alpha})")

            if beta <= alpha:
                print(f"Depth {depth}: PRUNING - beta ({beta}) <= alpha ({alpha}). Pruning remaining branches.")
                break # Alpha-Beta Pruning
        return min_eval
```

### Question 2

```python
def elevation_function(x):
    return -(x - 3)**2 + 9
```

```python
def hill_climbing():
    current_position = 0
    positions_visited = [current_position]

    print(f"Starting Hill Climbing from x = {current_position}")
    print(f"Initial Elevation at x = {current_position}: {elevation_function(current_position)}")

    while True:
        # Evaluate neighbors
        left_neighbor = current_position - 1
        right_neighbor = current_position + 1

        current_elevation = elevation_function(current_position)

        best_next_position = current_position
        best_next_elevation = current_elevation

        # Check left neighbor
        if 0 <= left_neighbor <= 6:
            left_elevation = elevation_function(left_neighbor)
            if left_elevation > best_next_elevation:
                best_next_elevation = left_elevation
                best_next_position = left_neighbor

        # Check right neighbor
        if 0 <= right_neighbor <= 6:
            right_elevation = elevation_function(right_neighbor)
            if right_elevation > best_next_elevation:
                best_next_elevation = right_elevation
                best_next_position = right_neighbor

        # If no better neighbor is found, we are at a local optimum
        if best_next_position == current_position:
            print(f"\nReached local optimum at x = {current_position} with elevation {current_elevation}")
            break
        else:
            current_position = best_next_position
            positions_visited.append(current_position)
            print(f"Moved to x = {current_position} with elevation {elevation_function(current_position)}")

    print(f"\nSequence of positions visited: {positions_visited}")
    print(f"Final position: {current_position}")
    print(f"Maximum elevation reached: {elevation_function(current_position)}")
```

```python
hill_climbing()
```

### Question 3

```python
def hill_climbing_resource_allocation():
    # productivity scores for each state
    productivity_scores = {
        'A': 3,
        'B': 6,
        'C': 5,
        'D': 7,
        'E': 9,
        'F': 8
    }

    # possible transitions between states (adjacency list)
    transitions = {
        'A': ['B', 'C'],
        'B': ['D'],
        'C': ['D'],
        'D': ['E', 'F'],
        'E': [],
        'F': []
    }

    current_state = 'A'
    visited_states = [current_state]

    print(f"Starting Hill Climbing from State: {current_state} (Productivity: {productivity_scores[current_state]})")

    while True:
        neighbors = transitions.get(current_state, [])

        if not neighbors:
            print(f"\nReached a state with no further transitions. Current state: {current_state}")
            break

        best_neighbor = None
        max_productivity_in_neighbors = productivity_scores[current_state]

        for neighbor in neighbors:
            neighbor_productivity = productivity_scores[neighbor]
            print(f"  Considering neighbor {neighbor} with productivity {neighbor_productivity}")
            if neighbor_productivity > max_productivity_in_neighbors:
                max_productivity_in_neighbors = neighbor_productivity
                best_neighbor = neighbor

        if best_neighbor and productivity_scores[best_neighbor] > productivity_scores[current_state]:
            print(f"Moving from {current_state} (Productivity: {productivity_scores[current_state]}) to {best_neighbor} (Productivity: {productivity_scores[best_neighbor]}) ")
            current_state = best_neighbor
            visited_states.append(current_state)
        else:
            print(f"\nNo better neighboring state found than {current_state} (Productivity: {productivity_scores[current_state]}).")
            break # local maximum

    print(f"\nSequence of states visited: {visited_states}")
    print(f"Final state reached: {current_state} with productivity {productivity_scores[current_state]}")
```

```python
hill_climbing_resource_allocation()
```

**Explanation for stopping at a local maximum:**

The Simple Hill Climbing algorithm stops when it reaches a state where none of its neighboring states have a higher productivity value. This state is considered a 'local maximum'. It's important to note that a local maximum is not necessarily the 'global maximum' (the highest possible productivity across all states). The algorithm only explores its immediate neighborhood and gets 'stuck' if it can't find a better path forward from its current position, even if a higher peak exists elsewhere that is not directly reachable via an increasing path from the current state.

In this specific problem, if the algorithm had started in a different state, it might have found a different, potentially higher, final productivity score. The algorithm does not backtrack or explore other branches once it commits to a path.
